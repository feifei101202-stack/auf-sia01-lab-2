## Setup

```python
import requests
import json

BASE_URL = 'https://fakestoreapi.com'

def pretty_print(response):
    print('Status Code:', response.status_code)
    try:
        print(json.dumps(response.json(), indent=4))
    except:
        print(response.text)
```

## PRODUCTS ENDPOINTS

```python
# Get all products
response = requests.get(f"{BASE_URL}/products")
pretty_print(response)
```

```python
# Get single product
response = requests.get(f"{BASE_URL}/products/1")
pretty_print(response)
```

```python
# Get limited products
response = requests.get(f"{BASE_URL}/products", params={'limit': 5})
pretty_print(response)
```

```python
# Get sorted products (desc)
response = requests.get(f"{BASE_URL}/products", params={'sort': 'desc'})
pretty_print(response)
```

```python
# Get categories
response = requests.get(f"{BASE_URL}/products/categories")
pretty_print(response)
```

```python
# Get products by category
response = requests.get(f"{BASE_URL}/products/category/electronics")
pretty_print(response)
```

```python
# Add new product
new_product = {
    'title': 'Lab Product',
    'price': 99.99,
    'description': 'Created during lab activity',
    'image': 'https://i.pravatar.cc',
    'category': 'electronics'
}
response = requests.post(f"{BASE_URL}/products", json=new_product)
pretty_print(response)
```

```python
# Update product
updated_product = {
    'title': 'Updated Lab Product',
    'price': 120.00,
    'description': 'Updated description',
    'image': 'https://i.pravatar.cc',
    'category': 'electronics'
}
response = requests.put(f"{BASE_URL}/products/1", json=updated_product)
pretty_print(response)
```

```python
# Delete product
response = requests.delete(f"{BASE_URL}/products/1")
pretty_print(response)
```

## CARTS ENDPOINTS

```python
# Get all carts
response = requests.get(f"{BASE_URL}/carts")
pretty_print(response)
```

```python
# Get single cart
response = requests.get(f"{BASE_URL}/carts/1")
pretty_print(response)
```

```python
# Get limited carts
response = requests.get(f"{BASE_URL}/carts", params={'limit': 3})
pretty_print(response)
```

```python
# Get carts in date range
params = {'startdate': '2019-12-10', 'enddate': '2020-10-10'}
response = requests.get(f"{BASE_URL}/carts", params=params)
pretty_print(response)
```

```python
# Get user carts
response = requests.get(f"{BASE_URL}/carts/user/2")
pretty_print(response)
```

```python
# Add new cart
new_cart = {
    'userId': 1,
    'date': '2026-02-24',
    'products': [
        {'productId': 1, 'quantity': 2},
        {'productId': 2, 'quantity': 1}
    ]
}
response = requests.post(f"{BASE_URL}/carts", json=new_cart)
pretty_print(response)
```

```python
# Update cart
updated_cart = {
    'userId': 1,
    'date': '2026-02-24',
    'products': [{'productId': 3, 'quantity': 5}]
}
response = requests.put(f"{BASE_URL}/carts/1", json=updated_cart)
pretty_print(response)
```

```python
# Delete cart
response = requests.delete(f"{BASE_URL}/carts/1")
pretty_print(response)
```
